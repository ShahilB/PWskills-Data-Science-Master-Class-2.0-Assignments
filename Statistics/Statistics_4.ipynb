## Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

### **Probability Mass Function (PMF) and Probability Density Function (PDF)**

The **Probability Mass Function (PMF)** and the **Probability Density Function (PDF)** are two fundamental concepts in probability theory used to describe discrete and continuous random variables, respectively.

---

### **1. Probability Mass Function (PMF)**
The **PMF** applies to **discrete** random variables. It gives the probability that a discrete random variable \(X\) takes a specific value \(x\). 

#### **Mathematical Definition:**
For a discrete random variable \(X\), the PMF is given by:
$$ 
P(X = x) = p(x)
 $$
where $$ p(x) $$ 
satisfies the following conditions:
1. $$ p(x) \geq 0 $$ for all possible values of x.
2. The sum of all probabilities is 1:
   $$
   \sum_{x} p(x) = 1
   $$

#### **Example of PMF:**
Consider a **fair six-sided die** where X  represents the outcome of rolling the die. The possible values of X are {1,2,3,4,5,6\} , and each occurs with equal probability:

$$
P(X = x) = \frac{1}{6}, \quad x \in \{1,2,3,4,5,6\}
$$ 

---

### **2. Probability Density Function (PDF)**
The **PDF** applies to **continuous** random variables. Instead of giving the probability of a single value (which is always 0 for continuous variables), it provides the probability **density** at any given point.

#### **Mathematical Definition:**
For a continuous random variable X, the PDF is denoted as  f(x) and satisfies:
1. $$ f(x) \geq 0 $$  for all  x .
2. The total probability across all values is 1:
   $$
   \int_{-\infty}^{\infty} f(x) dx = 1
   $$
3. The probability that X falls within a certain range [a, b] is given by:
   $$
   P(a \leq X \leq b) = \int_{a}^{b} f(x) dx
   $$

#### **Example of PDF:**
Consider a **normal distribution** with mean $$ \mu = 0 $$ and standard deviation $$ \sigma = 1 $$. The PDF is given by:

$$
f(x) = \frac{1}{\sqrt{2\pi}} e^{-x^2/2}, \quad -\infty < x < \infty
$$

Unlike PMF, the probability of getting an exact value (e.g., P(X = 0) is **zero** because for continuous distributions, probability is only meaningful over an interval.

---

### **Key Differences Between PMF and PDF**
| Feature  | PMF (Discrete) | PDF (Continuous) |
|----------|---------------|------------------|
| Type of variable | Discrete | Continuous |
| Definition | Probability at specific values | Probability density over a range |
| Sum/Integral | sum p(x) = 1 | int f(x)dx = 1 |
| Probability of exact value | P(X = x)  is nonzero |  P(X = x) = 0 , only intervals matter |


## Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

### **Cumulative Distribution Function (CDF)**

The **Cumulative Distribution Function (CDF)** is a function that gives the probability that a random variable X takes on a value **less than or equal to** a given number x. It is applicable to both **discrete** and **continuous** random variables.

---

### **Mathematical Definition:**
For a random variable X, the CDF is defined as:

$$ 
F(x) = P(X \leq x)
$$ 

where:
- F(x)  represents the cumulative probability up to  x .
-  X  can be either **discrete** or **continuous**.

---

### **CDF for Discrete Random Variables**
For a **discrete** random variable, the CDF is the sum of probabilities of all values **less than or equal to x**:

$$
F(x) = \sum_{t \leq x} P(X = t)
$$

#### **Example (Discrete Case):**
Consider rolling a fair six-sided die where X represents the outcome.

The PMF is:
$$
P(X = x) = \frac{1}{6}, \quad x \in \{1,2,3,4,5,6\}
$$

The CDF is calculated as:

$$
F(x) =
\begin{cases} 
0, & x < 1 \\
\frac{1}{6}, & 1 \leq x < 2 \\
\frac{2}{6}, & 2 \leq x < 3 \\
\frac{3}{6}, & 3 \leq x < 4 \\
\frac{4}{6}, & 4 \leq x < 5 \\
\frac{5}{6}, & 5 \leq x < 6 \\
1, & x \geq 6
\end{cases}
$$ 

This means, for example, \( F(3) = 3/6 = 0.5 \), meaning there is a **50% probability that the outcome is 3 or less**.

---

### **CDF for Continuous Random Variables**
For a **continuous** random variable, the CDF is found by integrating the **Probability Density Function (PDF)**:

$$
F(x) = \int_{-\infty}^{x} f(t) dt
$$

where \( f(x) \) is the **PDF** of \( X \).

#### **Example (Continuous Case):**
Consider a **normal distribution** $$ X \sim N(0,1) $$ with mean $$ \mu = 0 $$ and standard deviation $$ \sigma = 1 $$. The PDF is:

$$
f(x) = \frac{1}{\sqrt{2\pi}} e^{-x^2/2}
$$

The CDF is:

$$
F(x) = \int_{-\infty}^{x} \frac{1}{\sqrt{2\pi}} e^{-t^2/2} dt
$$

Since this integral does not have a simple closed-form solution, we use statistical tables or software to compute CDF values for a normal distribution.

For example:
-  F(0) = 0.5 (50% probability that $$ X \leq 0 $$ )
-  F(1) â‰ˆ 0.8413  (84.13% probability that $$ X \leq 1 $$ )

---

### **Why is CDF Used?**
1. **To Compute Probabilities Easily:**  
   - Instead of integrating the PDF manually, we can use the CDF to find probabilities directly.
   - Example: $$ P(a \leq X \leq b) = F(b) - F(a) $$.

2. **To Analyze Distribution Behavior:**  
   - CDFs provide a **cumulative** view of probabilities, which helps understand how values are distributed.

3. **For Statistical Applications:**  
   - CDFs are used in **hypothesis testing**, **percentile calculations**, and **quantile functions**.
   - Example: The median of a distribution is the value  x  where F(x) = 0.5 .

4. **For Generating Random Samples:**  
   - Many simulation and machine learning models use the **inverse transform method**, which relies on the CDF.

---

### **Key Differences Between PMF/PDF and CDF**
| Feature | PMF (Discrete) | PDF (Continuous) | CDF (Both) |
|---------|---------------|------------------|------------|
| Type of variable | Discrete | Continuous | Both |
| Definition | P(X = x)  | f(x) , density |  P(X <= x)  |
| Summation/Integration | sum P(X = x) | int f(x)dx | sum  or int  |
| Probability of exact value |  P(X = x) not equal to 0 ) | P(X = x) = 0 | Not applicable |
| Usage | Specific probabilities | Density estimation | Cumulative probability |

---

### **Conclusion**
- **CDF** gives the cumulative probability up to a certain value.
- It helps in computing probabilities over ranges and analyzing distributions.
- It is used for statistical analysis, simulations, and data modeling.

## Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

### **Bernoulli Distribution**  

The **Bernoulli distribution** is a discrete probability distribution for a random variable that has only **two possible outcomes**, typically labeled as **success (1)** and **failure (0)**. It is the simplest probability distribution used to model binary events.

---

### **Mathematical Definition:**  
A random variable (X) follows a **Bernoulli distribution** if it takes values **1 (success)** with probability p and **0 (failure)** with probability (1 - p).

$$
P(X = x) =
\begin{cases} 
p, & \text{if } x = 1 \\
1 - p, & \text{if } x = 0
\end{cases}
$$

where $$ 0 \leq p \leq 1 $$.

The **mean** (expected value) and **variance** of a Bernoulli-distributed random variable X are:  

- **Mean:**  E[X] = p   
- **Variance:** Var(X) = p(1 - p)   

---

### **Example of Bernoulli Distribution**  
- **Coin Toss**: If we flip a **fair coin**, the probability of getting heads (X = 1) is (p = 0.5), and the probability of getting tails (X = 0) is ( 1 - p = 0.5 ).
- **Passing a Test**: If a student has a 70% chance of passing an exam, we can model the outcome using a Bernoulli distribution with ( p = 0.7 ).

---

### **Difference Between Bernoulli Distribution and Binomial Distribution**  

| Feature | **Bernoulli Distribution** | **Binomial Distribution** |
|---------|------------------------|------------------------|
| Type | Single trial | Multiple trials |
| Number of Trials | ( n = 1 ) | ( n >= 1 ) |
| Random Variable | Single outcome (0 or 1) | Number of successes in  n  trials |
| Probability Mass Function (PMF) | $$ P(X = x) = p^x(1 - p)^{1 - x}$$  | $$ P(X = k) = \binom{n}{k} p^k (1 - p)^{n-k} $$|
| Example | Flipping a coin once | Flipping a coin 10 times and counting heads |

---

### **Example of Binomial Distribution**  
If we **flip a fair coin 10 times**, the number of heads follows a **binomial distribution** with:
- n = 10  (trials)
- p = 0.5  (probability of heads)
- The probability of getting exactly **5 heads** is:

$$
P(X = 5) = \binom{10}{5} (0.5)^5 (0.5)^5 = 0.246
$$

Thus, the **Bernoulli distribution is a special case of the binomial distribution** when  n = 1.


## Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.

### **Given Information:**
- Mean (Î¼) = 50  
- Standard Deviation (Ïƒ) = 10  
- We need to find **P(X > 60)**, assuming a normal distribution.

---

### **Step 1: Convert to Standard Normal Form (Z-score)**
The **Z-score** formula is:

$$
Z = \frac{X - \mu}{\sigma}
$$

Substituting the given values:

$$
Z = \frac{60 - 50}{10} = \frac{10}{10} = 1
$$

---

### **Step 2: Find the Probability Using Z-Table**
The **Z-table** provides the probability $$ P(X \leq 60) $$, which is the area to the left of  Z = 1.

From the **Z-table**, the probability of $$ Z \leq 1.00 $$ is:

$$
P(Z \leq 1) = 0.8413
$$

---

### **Step 3: Compute ( P(X > 60) )**
Since we need $$ P(X > 60) $$, we subtract the cumulative probability from 1:

$$
P(X > 60) = 1 - P(X \leq 60)
$$

$$
P(X > 60) = 1 - 0.8413 = 0.1587
$$

---

### **Final Answer:**
$$
P(X > 60) = 0.1587 \quad \text{or} \quad 15.87\%
$$

This means there is a **15.87% probability** that a randomly selected observation from this dataset will be **greater than 60**.



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Given parameters
mu = 50  # Mean
sigma = 10  # Standard Deviation
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)  # Range for x-axis

# Normal distribution curve
pdf = stats.norm.pdf(x, mu, sigma)

# Plot the normal distribution
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label="Normal Distribution", color='blue')

# Shade the area for P(X > 60)
x_fill = np.linspace(60, mu + 4*sigma, 100)
y_fill = stats.norm.pdf(x_fill, mu, sigma)
plt.fill_between(x_fill, y_fill, color='red', alpha=0.5, label="P(X > 60)")

# Add mean and threshold lines
plt.axvline(mu, color='black', linestyle="--", label="Mean (50)")
plt.axvline(60, color='red', linestyle="--", label="X = 60")

# Labels and title
plt.title("Normal Distribution with P(X > 60) Shaded")
plt.xlabel("X values")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()

# Show plot
plt.show()
```

Here's the visualization of the normal distribution with the probability P(X>60) shaded in red. This represents the 15.87% probability of randomly selecting an observation greater than 60.

## Q7: Explain uniform Distribution with an example.

### **Uniform Distribution**  

A **uniform distribution** is a type of probability distribution where all outcomes are equally likely. It is often used when there is no bias or preference for any particular outcome.  

There are **two types** of uniform distributions:  
1. **Discrete Uniform Distribution** â€“ A finite number of outcomes with equal probabilities.  
2. **Continuous Uniform Distribution** â€“ Any value within a given range is equally likely.

---

### **1. Discrete Uniform Distribution**
A random variable  X follows a **discrete uniform distribution** if it takes **n** different values, each with equal probability:

$$
P(X = x) = \frac{1}{n}, \quad \text{for } x = x_1, x_2, ..., x_n.
$$

#### **Example:**  
- Rolling a **fair die** (1 to 6)  
  - Each face has an equal probability of occurring:  
    $$
    P(X = x) = \frac{1}{6}, \quad x \in \{1,2,3,4,5,6\}
    $$

---

### **2. Continuous Uniform Distribution**
A random variable  X  follows a **continuous uniform distribution** if its probability density function (PDF) is:

$$
f(x) =
\begin{cases}
\frac{1}{b - a}, & \text{if } a \leq x \leq b \\
0, & \text{otherwise}
\end{cases}
$$

where:  
-  a  = lower bound  
-  b  = upper bound  

The **mean** and **variance** of a continuous uniform distribution are:  

$$
E[X] = \frac{a + b}{2}, \quad Var(X) = \frac{(b - a)^2}{12}
$$

#### **Example:**  
- **Bus Arrival Time**  
  Suppose a bus arrives **randomly** between **10:00 AM and 10:30 AM**. If we assume the bus is equally likely to arrive at any time within this period, the arrival time follows a **continuous uniform distribution** from **a = 0 minutes** to **b = 30 minutes**.  
  - The probability of the bus arriving between **10:10 and 10:20** is:  

    $$
    P(10 \leq X \leq 20) = \frac{20 - 10}{30 - 0} = \frac{10}{30} = 0.3333 \text{ (or 33.33%)}
    $$

---

### **Key Differences Between Uniform and Normal Distributions**
| Feature | **Uniform Distribution** | **Normal Distribution** |
|---------|------------------------|------------------------|
| Shape | Rectangular | Bell-shaped |
| Probability | All outcomes equally likely | Center values more likely |
| Mean | $$ frac{a + b}{2} $$ | $$\mu$$ (mean) |
| Spread | Constant probability | Defined by standard deviation ($$ \sigma $$ )|

```python
# Generate data for Uniform Distribution
a, b = 0, 30  # Lower and Upper bounds
x_uniform = np.linspace(a - 5, b + 5, 1000)  # X values for plot
y_uniform = np.where((x_uniform >= a) & (x_uniform <= b), 1/(b - a), 0)  # PDF

# Plot the uniform distribution
plt.figure(figsize=(8, 5))
plt.plot(x_uniform, y_uniform, color='blue', label="Uniform Distribution (0,30)")

# Shade the probability between 10 and 20
x_fill = np.linspace(10, 20, 100)
y_fill = np.full_like(x_fill, 1/(b - a))
plt.fill_between(x_fill, y_fill, color='red', alpha=0.5, label="P(10 â‰¤ X â‰¤ 20)")

# Add lines for bounds and shaded region
plt.axvline(a, color='black', linestyle="--", label="a = 0")
plt.axvline(b, color='black', linestyle="--", label="b = 30")
plt.axvline(10, color='red', linestyle="--", label="X = 10")
plt.axvline(20, color='red', linestyle="--", label="X = 20")

# Labels and title
plt.title("Continuous Uniform Distribution (0, 30)")
plt.xlabel("X values (Minutes)")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()

# Show plot
plt.show()
```

Hereâ€™s the visualization of a **continuous uniform distribution** from **0 to 30 minutes**. The red shaded region represents the probability of the event **(bus arriving between 10 and 20 minutes, which is 33.33%)**.  

## Q8: What is the z score? State the importance of the z score.

### **What is the Z-Score?**  
The **Z-score** (also called the **standard score**) measures how many **standard deviations** a data point is from the **mean** of a dataset. It helps in comparing individual values within different distributions.  

#### **Formula for Z-Score:**  
$$
Z = \frac{X - \mu}{\sigma}
$$
where:  
- X = observed value  
- Î¼ = mean of the dataset  
- Ïƒ = standard deviation  

A **positive Z-score** means the value is **above the mean**, while a **negative Z-score** means the value is **below the mean**.

---

### **Importance of Z-Score**
1. **Standardization of Different Data Distributions**  
   - Converts different datasets into a common scale, making comparisons easier.  
   - Example: Comparing student scores from different exams with different means and standard deviations.

2. **Identifying Outliers**  
   - If a Z-score is **greater than +3 or less than -3**, it is considered an **outlier** (rare event).  
   - Example: In a dataset of human heights, a Z-score of 4 would indicate an extremely tall person.

3. **Probability Calculations in Normal Distribution**  
   - Z-scores help find the probability of an event occurring in a **normal distribution** using **Z-tables**.  
   - Example: Finding the probability of a test score being greater than 85 when the mean is 70 and standard deviation is 10.

4. **Hypothesis Testing**  
   - Used in **Z-tests** to determine if sample data significantly deviates from the population mean.  
   - Example: Checking if a new drug significantly changes blood pressure levels.

5. **Machine Learning & Data Preprocessing**  
   - In ML, Z-score normalization (standardization) helps in handling datasets with different scales.  
   - Example: Standardizing features like age, income, and weight to improve model performance.

---

### **Example Calculation**
Suppose we have:
- Mean exam score: Î¼ = 70 
- Standard deviation: Ïƒ = 10 
- A student's score: X = 85 

**Z-score Calculation:**  
$$
Z = \frac{85 - 70}{10} = \frac{15}{10} = 1.5
$$
This means the studentâ€™s score is **1.5 standard deviations above the mean**.

```python
# Generate data for Normal Distribution
mu = 70  # Mean
sigma = 10  # Standard Deviation
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)  # X values for plot
pdf = stats.norm.pdf(x, mu, sigma)  # Normal distribution

# Plot the normal distribution
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label="Normal Distribution (Î¼=70, Ïƒ=10)", color='blue')

# Shade the area for Z = 1.5 (X > 85)
x_fill = np.linspace(85, mu + 4*sigma, 100)
y_fill = stats.norm.pdf(x_fill, mu, sigma)
plt.fill_between(x_fill, y_fill, color='red', alpha=0.5, label="Z > 1.5 (X > 85)")

# Add mean and threshold lines
plt.axvline(mu, color='black', linestyle="--", label="Mean (70)")
plt.axvline(85, color='red', linestyle="--", label="X = 85 (Z = 1.5)")

# Labels and title
plt.title("Z-Score Visualization for X = 85")
plt.xlabel("X values (Exam Scores)")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()

# Show plot
plt.show()
```

Hereâ€™s the visualization of the Z-score on a normal distribution curve. The red-shaded area represents the probability of a score greater than 85 (Z = 1.5).

## Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

### **What is the Central Limit Theorem (CLT)?**  
The **Central Limit Theorem (CLT)** states that when you take a large number of **random samples** from any population (regardless of its original distribution), the **sampling distribution of the sample mean** will be **approximately normal** if the sample size is sufficiently large (\( n \geq 30 \)).

#### **Mathematical Formulation**  
If we take random samples of size  n  from a population with:  
- Mean (Î¼)  
- Standard deviation (Ïƒ)  

Then, the **sampling distribution of the sample mean** xÌ„ follows:  
$$
\bar{X} \sim N \left( \mu, \frac{\sigma}{\sqrt{n}} \right)
$$
where:  
- $$ \mu_{\bar{X}} = \mu $$ (Mean of sample means remains the same as the population mean)  
- $$ \sigma_{\bar{X}} = \frac{\sigma}{\sqrt{n}} $$ (Standard deviation of sample means, called **standard error**)  

Thus, as **sample size n increases**, the distribution of 'xÌ„' **approaches a normal distribution** even if the original population is **not normal**.

---

### **Significance of the Central Limit Theorem**
1. **Foundation of Inferential Statistics**  
   - CLT enables us to use the **normal distribution** to make **confidence intervals** and **hypothesis tests**, even if the population distribution is unknown.

2. **Justifies the Use of the Normal Distribution**  
   - Many real-world problems rely on **sample means**, and thanks to CLT, we can assume normality for large samples.

3. **Estimation of Population Parameters**  
   - Helps estimate the **population mean and standard deviation** using sample statistics.

4. **Improves Accuracy with Larger Samples**  
   - As **sample size increases**, the standard error **decreases**, making estimates more **precise**.

5. **Useful in Business, Finance, and Machine Learning**  
   - **Example:** In stock market analysis, the CLT helps approximate the average return of stocks using sample data.

---

### **Example of CLT in Action**  
**Scenario:** Suppose the population distribution of test scores is skewed.  
- We take **random samples** of **size 30 or more** and calculate their **means**.  
- As we take more samples, the **distribution of sample means** becomes **approximately normal**.


```python

# Re-import necessary libraries since execution state was reset
import numpy as np
import matplotlib.pyplot as plt

# Simulating Central Limit Theorem

# Generate a highly skewed population distribution (Exponential Distribution)
population = np.random.exponential(scale=2, size=10000)  # Population of 10,000

# Sample sizes to compare
sample_sizes = [5, 30, 100]

# Plot setup
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

# Plot original population distribution (highly skewed)
axes[0].hist(population, bins=30, color='blue', alpha=0.7, density=True)
axes[0].set_title("Original Skewed Population")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Density")

# Plot distributions of sample means for different sample sizes
for i, n in enumerate(sample_sizes):
    sample_means = [np.mean(np.random.choice(population, n)) for _ in range(1000)]
    axes[i + 1].hist(sample_means, bins=30, color='red', alpha=0.7, density=True)
    axes[i + 1].set_title(f"Sample Size = {n}")
    axes[i + 1].set_xlabel("Sample Mean")
    axes[i + 1].set_ylabel("Density")

plt.tight_layout()
plt.show()
```

Hereâ€™s a simulation of the Central Limit Theorem (CLT) in action:

First plot (Leftmost): The original population follows a skewed exponential distribution.
Next three plots: As we increase the sample size from 5 â†’ 30 â†’ 100, the sample mean distribution becomes more normal, even though the original population was not normal.
This demonstrates how CLT ensures that the sampling distribution of the mean approaches normality as sample size increases!

## Q10: State the assumptions of the Central Limit Theorem.

### **Assumptions of the Central Limit Theorem (CLT)**  

For the **Central Limit Theorem (CLT)** to hold, the following assumptions must be satisfied:  

1. **Random Sampling**  
   - The samples must be drawn randomly and independently from the population.  
   - **Why?** Ensures unbiased representation of the population.  

2. **Independence of Observations**  
   - The selected samples should be **independent** of each other.  
   - If sampling **without replacement**, the sample size **should not exceed 10%** of the population (\( n \leq 0.1N \)) to maintain near-independence.  
   - **Why?** Prevents correlation between observations that may distort results.  

3. **Sample Size Should Be Sufficiently Large**  
   - A **larger sample size (\( n \geq 30 \))** is generally needed for the sampling distribution of the mean to be approximately normal.  
   - **Why?** Larger samples reduce variability and improve normality.  

4. **Finite Variance and Mean**  
   - The population from which the sample is drawn should have a **finite mean (\( \mu \))** and **finite variance (\( \sigma^2 \))**.  
   - **Why?** Extreme values (infinite variance) can distort the sampling distribution.  

5. **If the Population is Normal, Any Sample Size Works**  
   - If the original population is **already normally distributed**, then the sampling distribution of the mean is also normal **regardless of sample size**.  
   - **Why?** CLT applies immediately without needing a large \( n \).  

---

### **Summary Table of CLT Assumptions**
| Assumption | Explanation |
|------------|------------|
| **Random Sampling** | The sample must be randomly selected. |
| **Independence** | Each observation must be independent of others. |
| **Sample Size ( n >= 30 )** | Large samples improve normality. |
| **Finite Mean & Variance** | The population must have a well-defined mean and variance. |
| **Normal Population Exception** | If the population is normal, CLT applies for any \( n \). |
