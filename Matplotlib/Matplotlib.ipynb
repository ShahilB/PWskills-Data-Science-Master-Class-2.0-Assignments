### ‚úÖ **Q1: What is Matplotlib? Why is it used?**

**Matplotlib** is a popular Python library used for **data visualization**. It allows users to create **static, animated, and interactive plots** in a variety of formats.

It is especially useful for:
- Understanding data patterns and trends
- Presenting results in a clear and visual format
- Debugging data and model behavior during analysis or machine learning

The `pyplot` module in `matplotlib` is commonly used and provides a **MATLAB-like interface** for creating plots.

---

### üîç **Why is it used?**
- To **visualize** data in the form of charts and graphs.
- To **communicate insights** derived from data.
- To **debug** and analyze relationships and patterns.

---

### üìä **Five Common Plots using `matplotlib.pyplot`:**

1. **Line Plot**
2. **Bar Plot**
3. **Histogram**
4. **Scatter Plot**
5. **Pie Chart**

---

### ‚úÖ Example (Using Pyplot):

````python
import matplotlib.pyplot as plt

# Example: Line plot
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Example Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
````

### ‚úÖ **Q2: What is a Scatter Plot?**

A **scatter plot** is a type of plot used to **display values for typically two variables** for a set of data. The data are displayed as a collection of points, where each point has:
- An **x-coordinate** (from the first variable)
- A **y-coordinate** (from the second variable)

Scatter plots are useful to:
- Detect **relationships** or **correlations** between variables
- Spot **outliers**
- Visualize **distribution patterns**

---

### ‚úÖ **Scatter Plot using the Given Code:**

````python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(3)

# Generate data
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create scatter plot
plt.scatter(x, y)

# Add labels and title
plt.title("Scatter Plot of Generated Data")
plt.xlabel("X Values")
plt.ylabel("Y Values")

# Show plot
plt.show()
````

---

### üñºÔ∏è **Explanation of Plot:**
- Each point on the scatter plot represents a randomly generated `(x, y)` pair.
- The points are scattered because of the **random normal distribution**.

Use the following data:

import numpy as np

### ‚úÖ **Q3: Why is the `subplot()` function used?**

The `subplot()` function in **Matplotlib** is used to **create multiple plots in a single figure**.

It lets you:
- Organize plots into a **grid layout** (rows √ó columns)
- Display **multiple charts side by side** or stacked
- Make comparisons between different datasets visually easier

---

### üß† **Syntax:**
```python
plt.subplot(nrows, ncols, index)
```
Where:
- `nrows` = number of rows in the figure
- `ncols` = number of columns
- `index` = position of the plot (starts at 1)

---

### ‚úÖ **Code: Four Line Plots using `subplot()`**

````python
import numpy as np
import matplotlib.pyplot as plt

# Data for all 4 lines
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.figure(figsize=(10, 8))

# Line 1
plt.subplot(2, 2, 1)
plt.plot(x, y1, color='blue')
plt.title("Line 1")

# Line 2
plt.subplot(2, 2, 2)
plt.plot(x, y2, color='green')
plt.title("Line 2")

# Line 3
plt.subplot(2, 2, 3)
plt.plot(x, y3, color='orange')
plt.title("Line 3")

# Line 4
plt.subplot(2, 2, 4)
plt.plot(x, y4, color='red')
plt.title("Line 4")

# Adjust layout
plt.tight_layout()
plt.show()
````

---

### üñºÔ∏è **What You‚Äôll See:**
- A 2√ó2 grid of plots.
- Each subplot has its own title and line data.
- All graphs are neatly arranged in one window.

### ‚úÖ **Q4: What is a Bar Plot? Why is it Used?**

A **bar plot** (or **bar chart**) is a graphical representation of **categorical data** using rectangular bars.  
Each bar's **length or height** represents the **value or frequency** of the corresponding category.

---

### üîç **Why is it used?**
- To **compare** data across different categories.
- To **visualize differences** or trends.
- Great for showing **discrete variables** (e.g., sales by company, votes by candidate, etc.)

---

### üìä **Code: Bar Plot & Horizontal Bar Plot**

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot (Vertical)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(company, profit, color='skyblue')
plt.title("Company Profits (Bar Plot)")
plt.xlabel("Company")
plt.ylabel("Profit ($ in millions)")

# Horizontal Bar Plot
plt.subplot(1, 2, 2)
plt.barh(company, profit, color='salmon')
plt.title("Company Profits (Horizontal Bar Plot)")
plt.xlabel("Profit ($ in millions)")
plt.ylabel("Company")

# Adjust layout
plt.tight_layout()
plt.show()
````

---

### üñºÔ∏è **What You'll See:**
- A **vertical bar chart** comparing company profits.
- A **horizontal bar chart** showing the same data from a different angle ‚Äî easier to read for long category names.

### ‚úÖ **Q5: What is a Box Plot? Why is it Used?**

A **box plot** (also called a **box-and-whisker plot**) is a statistical plot that **shows the distribution** of numerical data through their **quartiles**.

---

### üì¶ **Key Components of a Box Plot:**
- **Box**: Spans from the 1st quartile (Q1) to the 3rd quartile (Q3)
- **Line in the box**: Represents the **median** (Q2)
- **Whiskers**: Extend to the minimum and maximum values within 1.5√ó IQR (interquartile range)
- **Dots (optional)**: Represent **outliers**

---

### üîç **Why is it used?**
- To **visualize the spread** and **central tendency** of data
- To **detect outliers**
- To compare **distributions** between datasets

---

### üìä **Code: Plotting a Box Plot**

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine data
data = [box1, box2]

# Create box plot
plt.figure(figsize=(8, 5))
plt.boxplot(data, labels=["Box1", "Box2"])
plt.title("Box Plot of Two Distributions")
plt.ylabel("Values")
plt.grid(True)
plt.show()
````

---

### üñºÔ∏è **What You‚Äôll See:**
- Two vertical box plots side by side
- Each shows median, spread, and outliers for `box1` and `box2`
- Great for comparing **variability** (box2 will likely show a wider box due to higher standard deviation)
