Q5. Explain Primary Key and Foreign Key.


Primary Key:

Definition: A primary key is a unique identifier for a record in a database table. It ensures that each record can be uniquely identified and prevents duplicate entries.
Characteristics:
Uniqueness: Each value in the primary key column(s) must be unique across the table.
Non-null: Primary key columns cannot contain NULL values.
Single or Composite: A primary key can consist of a single column or a combination of columns (composite primary key).
Example: In a table named Employees, the EmployeeID column could be the primary key.

CREATE TABLE Employees (
    EmployeeID INT PRIMARY KEY,
    FirstName VARCHAR(50),
    LastName VARCHAR(50),
    HireDate DATE
);

Foreign Key:
Definition: A foreign key is a field (or collection of fields) in one table that uniquely identifies a row of another table. The foreign key establishes a relationship between the two tables.
Characteristics:
Referential Integrity: Ensures that the value in the foreign key column matches a value in the primary key column of the referenced table.
Constraints: Helps maintain data integrity by enforcing valid relationships between tables.
Nullable or Non-nullable: Foreign key columns can contain NULL values if the relationship is optional.
Example: In a table named Orders, the CustomerID column could be a foreign key referencing the CustomerID in the Customers table.

CREATE TABLE Orders (
    OrderID INT PRIMARY KEY,
    OrderDate DATE,
    CustomerID INT,
    FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID)
);

Summary:

Primary Key: Ensures each record in a table is unique and identifiable. It cannot have duplicate or NULL values.
Foreign Key: Creates a link between two tables by referencing the primary key of another table, ensuring referential integrity and enabling relationships between tables.

These keys are fundamental in relational database design, enabling efficient data retrieval, maintaining data integrity, and enforcing relationships between different data entities.


Q6. Write a python code to connect MySQL to python. Explain the cursor() and execute() method.


```python
import mysql.connector

# Establish a connection to the database
connection = mysql.connector.connect(
    host='localhost',      # Database host
    user='your_username',  # Database username
    password='your_password',  # Database password
    database='your_database'  # Database name
)

# Create a cursor object using the cursor() method
cursor = connection.cursor()

# Execute a query using the execute() method
cursor.execute("SELECT * FROM Employees")

# Fetch all the rows from the executed query
rows = cursor.fetchall()

# Print the fetched rows
for row in rows:
    print(row)

# Close the cursor and connection
cursor.close()
connection.close()
```

Explanation:

cursor() Method:

Purpose: The cursor() method creates a cursor object that allows you to execute SQL queries and fetch data from the database.

Usage:

cursor = connection.cursor()

Function: This line creates a new cursor object named cursor using the connection object.

execute() Method:

Purpose: The execute() method is used to execute a SQL query through the cursor object.
Usage:

cursor.execute("SELECT * FROM Employees")

Function: This line executes the SQL query "SELECT * FROM Employees". The query is sent to the database server for execution.

Detailed Steps:

Import the mysql.connector Library:

The library is imported to enable MySQL database interactions in Python.
Establish a Connection:

The mysql.connector.connect method is used to establish a connection to the MySQL database by providing the necessary credentials and database details.

connection = mysql.connector.connect(
    host='localhost',
    user='your_username',
    password='your_password',
    database='your_database'
)

Create a Cursor:

The cursor() method of the connection object is called to create a cursor. This cursor will be used to execute SQL queries.

cursor = connection.cursor()
Execute a Query:

The execute() method of the cursor object is used to run an SQL query. Here, it retrieves all records from the Employees table.

cursor.execute("SELECT * FROM Employees")
Fetch and Process Data:

The fetchall() method retrieves all rows from the executed query, which are then printed.

rows = cursor.fetchall()
for row in rows:
    print(row)
    
Close the Cursor and Connection:

It’s good practice to close the cursor and the database connection after operations are complete to free up resources.

cursor.close()
connection.close()

This example demonstrates how to connect to a MySQL database, execute a query, fetch results, and handle database resources properly using Python.
